```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
%matplotlib inline
```

```python
t = np.linspace(0, 20, 101)
x = np.cos(-t)
y = np.cos(-t)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(t, x, y)
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_zlabel('y')
ax.set_xticks(np.linspace(0, 20, 5))
ax.set_zlim(-2,2)
ax.view_init(30, -30)
plt.title(r"$E_0' = E_0, \phi = 0$")
plt.tight_layout()
plt.savefig("1.pdf")
```

```python
t = np.linspace(0, 20, 101)
x = np.cos(-t)
y = np.cos(-t + pi/4)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(t, x, y)
ax.set_xlabel('t')
ax.set_ylabel('x')
ax.set_zlabel('y')
ax.set_xticks(np.linspace(0, 20, 5))
# ax.set_yticklabels([])
ax.set_zlim(-2,2)
# ax.set_zticklabels([])
ax.view_init(30, -30)
plt.title(r"$E_0' = E_0, \phi = \pi/4$")
plt.tight_layout()
plt.savefig("2.pdf")
```
