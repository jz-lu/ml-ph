```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
b = np.load("/Users/jonathanlu/Documents/tvGr/b.npy")
bx, by = b.T
energies = np.load("/Users/jonathanlu/Documents/tvGr/e.npy")
```

```python
fig, ax = plt.subplots()
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
plt.rc('font', size=16)
cf = ax.tricontourf(bx, by, energies-0.04, levels=501, cmap="RdGy")
plt.colorbar(cf, ax=ax, shrink=0.6, ticks=[0,5,10,15,20])
ax.set_xlabel(r"$b_x$ $(\mathrm{\AA})$", fontsize=18)
ax.set_ylabel(r"$b_y$ $(\mathrm{\AA})$", fontsize=18)
ax.set_aspect('equal') # prevent axis stretching
plt.tight_layout()
plt.savefig('/Users/jonathanlu/Documents/data_mlph/GSFE.png')
```
