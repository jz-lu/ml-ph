# Comparison with MD calculations at commensurate angles
Part 2: bands

```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
%matplotlib inline
```

```python
# Load the displacements, indexed by atomic index
data = np.loadtxt('/Users/jonathanlu/Documents/tvGr/MD/Phonon_dispersion.dat')
x, y = data.T
y *= 8.066 # meV -> 1/cm
```

```python
fig, ax = plt.subplots(figsize=(8,6))

# make the default Font latex
plt.rc('font', size=28)
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'

plt.scatter(x, y, c='k', s=6)
plt.xlim(np.min(x), np.max(x))
plt.xticks([0, 60, 120, 180], ['K', '$\Gamma$', 'M', 'K'])
plt.ylim(0, 100)
plt.ylabel(r"$\omega$ (cm$^{-1}$)")
# plt.title(r"$\theta$ = $7.39^\circ$")
plt.tight_layout()
plt.savefig("/Users/jonathanlu/Documents/data_mlph/MD_band.pdf")
```
