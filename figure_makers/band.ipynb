```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
THETA = 7.39
evals = np.load(f"/Users/jonathanlu/Documents/tvGr/evals/evals_{THETA}.npy")
evals.shape
```

```python
signs = (evals < 0)*(-1) + (evals >= 0)*1
modes = np.sqrt(np.abs(evals)) * signs
print(f"Minimum mode: {np.min(modes)}")

# make the default Font latex
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'

_, ax = plt.subplots()
plt.rc('font', size=22)
for i, L_set in enumerate(modes):
    plt.scatter([i]*len(L_set), L_set, c='k', s=0.7)
plt.ylim(bottom=np.min(modes), top=30)
plt.xlim(0, len(evals)-1)

plt.ylabel(r"$\omega$ (cm$^{-1}$)")
plt.xticks([0, 100, 200, 300], ['K', r'$\Gamma$', 'M', 'K'])
# plt.xticks([0, 100, 200, 300], labels=[])
plt.yticks([0,30,60,90,120])

props = dict(boxstyle='round', facecolor='lightblue', alpha=1)
# place a text box in upper left in axes coords
# plt.text(0.84, 0.12, fr'{THETA}$^\circ$', transform=ax.transAxes, fontsize=22,
#         verticalalignment='top', bbox=props)
plt.tight_layout()
plt.savefig(f"/Users/jonathanlu/Documents/data_mlph/Gr_band_{THETA}.pdf")
plt.show()
```
