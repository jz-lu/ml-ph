```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
THETA = 8.0
evals = np.load(f"/Users/jonathanlu/Documents/tvGr/evals/evals_{THETA}.npy")
evals.shape
```

```python
signs = (evals < 0)*(-1) + (evals >= 0)*1
modes = np.sqrt(np.abs(evals)) * signs
print(f"Minimum mode: {np.min(modes)}")

# make the default Font latex
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'

_, ax = plt.subplots()
FONT_SIZE = 28
plt.rc('font', size=FONT_SIZE)
for i, L_set in enumerate(modes):
    plt.scatter([i]*len(L_set), L_set, c='grey', s=0.7)
plt.ylim(bottom=-10, top=30)
plt.xlim(0, len(evals)-1)

# plt.ylabel(r"$\omega$ (cm$^{-1}$)", fontsize=FONT_SIZE)
plt.xticks([0, 100, 200, 300], ['K', r'$\Gamma$', 'M', 'K'], fontsize=FONT_SIZE)
# plt.xticks([0, 100, 200, 300], labels=[])
plt.yticks([0,50,100,150], fontsize=FONT_SIZE)
plt.plot([100]*2, ax.get_ylim(), c='k')
plt.plot([200]*2, ax.get_ylim(), c='k')

plt.tight_layout()
plt.savefig(f"/Users/jonathanlu/Documents/data_mlph/Gr_band_{THETA}.pdf")
plt.show()
```
