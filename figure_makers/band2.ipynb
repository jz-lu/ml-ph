```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
```

```python
THETA = 0.4
evals = np.load(f"/Users/jonathanlu/Documents/tvGr/evals/evals_{THETA}.npy")
signs = (evals < 0)*(-1) + (evals >= 0)*1
modes = np.sqrt(np.abs(evals)) * signs
print(f"Minimum mode: {np.min(modes)}")
```

```python
# make the default Font latex
mpl.rcParams['mathtext.fontset'] = 'stix'
mpl.rcParams['font.family'] = 'STIXGeneral'

HIGH = 30
LOW = -5

FONT_SIZE = 28
plt.rc('font', size=18)
fig, ax = plt.subplots(figsize=(4,5.5))
for i, L_set in enumerate(modes):
    plt.scatter([i]*len(L_set), L_set, c='grey', s=0.5)
plt.ylim(bottom=LOW, top=HIGH)
plt.xlim(0, len(modes)-1)

# plt.ylabel(r"$\omega^2$ (cm$^{-2}$)", fontsize=FONT_SIZE)
# plt.xticks([0, 100, 200, 300], ['K', r'$\Gamma$', 'M', 'K'], fontsize=FONT_SIZE-2)
plt.xticks([0, 100, 200, 300], labels=[])
plt.yticks([0, 10, 20, 30], fontsize=FONT_SIZE-6)
plt.plot([100]*2, [LOW, HIGH], c='k', linewidth=1)
plt.plot([200]*2, [LOW, HIGH], c='k', linewidth=1)

plt.tight_layout()
plt.savefig(f"/Users/jonathanlu/Documents/data_mlph/Gr_band_{THETA}.pdf")
plt.show()
```
